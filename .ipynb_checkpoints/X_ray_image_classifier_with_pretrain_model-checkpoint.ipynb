```python
# import the library 
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
import os 
```

```python
DataDir =r"../chest_xray/chest_xray/train"
```

```python
#installizing the category of classification
CATEGORIES =["NORMAL","PNEUMONIA"] 
```

```python
for i in CATEGORIES: 
    path =os.path.join(DataDir,i) 
    for img in os.listdir(path):
        img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
        plt.imshow(img_array,cmap='gray') 
        plt.show()
        break
    break
```

```python
## intializing the size of the image so that every image should be in same dimension 
img_size =100
```

```python
#image after resize and gray scale 
new_array =cv2.resize(img_array,(img_size,img_size)) 
plt.imshow(new_array,cmap='gray') 
plt.show()
```

```python
training_data =[]
```

```python
# defining the function for training data according to the categories converting the image into gray scale 
# converting it into numpy arrays 
def create_training_data(): 
    for i in CATEGORIES:

        path =os.path.join(DataDir,i) 
        class_num =CATEGORIES.index(i) 
        for img in os.listdir(path):
            try: 
                img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
                new_array =cv2.resize(img_array,(img_size,img_size)) 
                training_data.append([new_array,class_num]) 

            except Exception as e:
                pass
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
import random 
random.shuffle(training_data) 
for sample in training_data[:10]:
    print(sample)
```

```python
x=[]
y=[]
```

```python
# spliting the feature and labels 
for features,label in training_data:
    x.append(features)
    y.append(label)
```

```python
print(x[0].reshape(-1,img_size,img_size,1)) 
```

```python
y=np.array(y) 
```

```python
#reshapping the feature for making it compatible with tensorflow 
x =np.array(x).reshape(-1,img_size,img_size,1) 
```

```python
validation_data =[] 
DataDir_val =r"../chest_xray/chest_xray/val"
```

```python
def create_validation_data(): 
    for i in CATEGORIES:

        path =os.path.join(DataDir_val,i) 
        class_num =CATEGORIES.index(i) 
        for img in os.listdir(path):
            try: 
                img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
                new_array =cv2.resize(img_array,(img_size,img_size)) 
                validation_data.append([new_array,class_num]) 

            except Exception as e:
                pass
```

```python
create_validation_data()
```

```python
len(validation_data)
```

```python
import random 
random.shuffle(validation_data) 
for sample in validation_data[:10]: 
    print(sample)
```

```python
x_val =[] 
y_val =[] 
```

```python
for features,label in validation_data: 
    x_val.append(features)
    y_val.append(label)
```

```python
y_val =np.array(y_val) 
x_val =np.array(x_val).reshape(-1,img_size,img_size,1)
```

```python
#import the library for training the model and adding the neural network layers 
import tensorflow as tf 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D 
import pickle 
```

```python
# pixels value ranges from 0 to 255 ,0 means black ,255 mean white ....
# In Order to scale it down to a Scale of 0 to 1 ,we are performing this step 
x=x/255.0
x_val=x_val/255.0
```

```python
x.shape
```

```python
x_val.shape
```

```python
## initialize the neural network layer for training the model 
model=Sequential() 
model.add(Conv2D(64,(3,3),input_shape=x.shape[1:])) 
model.add(Activation('relu')) 
model.add(MaxPooling2D(pool_size=(2,2))) 
model.add(Dropout(.2)) 

model.add(Conv2D(128,(3,3))) 
model.add(Activation('relu')) 
model.add(MaxPooling2D(pool_size=(2,2))) 
model.add(Dropout(.2))

model.add(Conv2D(128,(3,3))) 
model.add(Activation('relu')) 
model.add(MaxPooling2D(pool_size=(2,2))) 
model.add(Dropout(.2)) 

model.add(Flatten()) 
model.add(Dense(64)) 

model.add(Dropout(0.5))
model.add(Dense(1)) 
model.add(Activation('sigmoid')) 
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,batch_size=4,epochs=10,validation_data=(x_val,y_val)) 
```

```python
model.save(r"../chest_xray/chest_xray/model_10.h5")
```

```python
# import the library 
import os 
import cv2 
import numpy as np 
import tensorflow as tf 
from tensorflow.keras.models import load_model  


# set the path for test data 
test_dir  ="../chest_xray/chest_xray/test" 

# define category 
CATEGORIES =["NORMAL","PNEUMONIA"] 

# set image size 
img_size =100 
model =load_model("../chest_xray/chest_xray/model_10.h5") 

# initialize lists for storing test data 
x_test =[] 
y_test =[] 

# loop through the best data directory and extract the images and their labels 
for category in CATEGORIES: 
    path =os.path.join(test_dir,category) 
    class_num =CATEGORIES.index(category) 
    for img in os.listdir(path):
        try: 
            img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
            new_array =cv2.resize(img_array,(img_size,img_size)) 
            x_test.append(new_array) 
            y_test.append(class_num) 
        except Exception as e: 
            pass 

#convert test data to numpy array 
x_test =np.array(x_test).reshape(-1,img_size,img_size,1) 
y_test =np.array(y_test) 

#normalize test data 
x_test =x_test/255.0 

# calculate test accuracy 
test_loss,test_acc =model.evaluate(x_test,y_test,verbose=2) 

print('test accuracy ',test_acc)
```

## Pre-train models

```python
# import necessary library
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# define image size
img_size = 100

# load the pre-trained model (VGG16)
base_model = VGG16(input_shape=(img_size, img_size, 3), include_top=False, weights='imagenet')

# freeze the layers of the pre-trained model
for layer in base_model.layers:
    layer.trainable = False

# add custom layers for classification
x = Flatten()(base_model.output)
x = Dense(256, activation='relu')(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

# create the new model
model = Model(inputs=base_model.input, outputs=predictions)

# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# define the image generators for training and validation data
train_datagen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
val_datagen = ImageDataGenerator(rescale=1./255)

# specify the training and validation data directories
train_dir = "../chest_xray/chest_xray/train"
val_dir = "../chest_xray/chest_xray/val"

# create the image generator for training data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_size, img_size),
    batch_size=30,
    class_mode='binary'
)

# create the image generator for validation data
val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(img_size, img_size),
    batch_size=30,
    class_mode='binary'
)

# train the model
model.fit(train_generator, epochs=10, validation_data=val_generator)

# evaluate the model on test data
test_dir = "../chest_xray/chest_xray/test"
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_size, img_size),
    batch_size=30,
    class_mode='binary'
)

test_loss, test_acc = model.evaluate(test_generator)
print("Test Accuracy:", test_acc)
```
