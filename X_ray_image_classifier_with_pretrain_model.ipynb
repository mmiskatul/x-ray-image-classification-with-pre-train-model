```python
# import the library 
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
import os 
```

```python
DataDir =r"../chest_xray/chest_xray/train"
```

```python
#installizing the category of classification
CATEGORIES =["NORMAL","PNEUMONIA"] 
```

```python
for i in CATEGORIES: 
    path =os.path.join(DataDir,i) 
    for img in os.listdir(path):
        img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
        plt.imshow(img_array,cmap='gray') 
        plt.show()
        break
    break
```

```python
## intializing the size of the image so that every image should be in same dimension 
img_size =100
```

```python
#image after resize and gray scale 
new_array =cv2.resize(img_array,(img_size,img_size)) 
plt.imshow(new_array,cmap='gray') 
plt.show()
```

```python
training_data =[]
```

```python
# defining the function for training data according to the categories converting the image into gray scale 
# converting it into numpy arrays 
def create_training_data(): 
    for i in CATEGORIES:

        path =os.path.join(DataDir,i) 
        class_num =CATEGORIES.index(i) 
        for img in os.listdir(path):
            try: 
                img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE) 
                new_array =cv2.resize(img_array,(img_size,img_size)) 
                training_data.append([new_array,class_num]) 

            except Exception as e:
                pass
```
